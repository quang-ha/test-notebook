```python
from ipywidgets import interact
import numpy as np

from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
output_notebook()
```

```python
# Define parameters
k = 10.0
m = 1.0
T = 10.0
dt = 0.001

# Create an array of time
time = np.linspace(0.0, T, T/dt)

# Initial state
x_0 = 0.0
v_0 = 1.0

# Using Euler for this initial setting
x_results = []
x = x_0
v = v_0

# Euler integration
for t in time:
    a = -(k/m)*x
    x = x + v*dt
    v = v + a*dt
    
    # Store results
    x_results.append(x)
```

```python
p = figure(title="simple line example", plot_height=300, plot_width=600, y_range=(-5,5))
r = p.line(time, x_results, color="#2222aa", line_width=3)
```

```python
def euler(k, m, dt):
    # Declare the initial condition
    x = x_0
    v = v_0
    x_results = []

    # Now start the integration
    for t in time:
        # Euler
        a = -(k/m)*x
        x = x + v*dt
        v = v + a*dt
    
        # Store results
        x_results.append(x)
    
    r.data_source.data['y'] = x_results
    push_notebook()
```

```python
show(p)
```

```python
interact(euler, k=(1,10), m=(1,10), dt=(0, 0.06, 0.001))
```
